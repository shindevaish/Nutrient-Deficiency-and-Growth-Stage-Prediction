```python
import os
import cv2

input_folder = "maize_growth_stage/flowering"
output_folder = "maize_growth_stage/flowering_zoomed"

os.makedirs(output_folder, exist_ok=True)

zoom_factor = 1.1

for file in os.listdir(input_folder):
    if file.lower().endswith(('.png', '.jpg', '.jpeg')):
        
        img_path = os.path.join(input_folder, file)
        img = cv2.imread(img_path)

        h, w = img.shape[:2]

        new_h = int(h / zoom_factor)
        new_w = int(w / zoom_factor)

        start_x = (w - new_w) // 2
        start_y = h - new_h
        cropped = img[start_y:start_y+new_h, start_x:start_x+new_w]
        zoomed = cv2.resize(cropped, (w, h))

        output_path = os.path.join(output_folder, "zoom_" + file)
        cv2.imwrite(output_path, zoomed)

print("Bottom-anchored zoom completed")
```
