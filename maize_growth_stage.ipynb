```python
import os
import pandas as pd
from PIL import Image

csv_file = "maize_growth_stage.csv"

source_folder = "images"

seedling_folder = "maize_growth_stage/seedling"

os.makedirs(seedling_folder, exist_ok=True)

df = pd.read_csv(csv_file)

seedling = df[
    (df["Maize Growth Stage 1"] == '1') | 
    (df["Maize Growth Stage 2"] == '1')
]

# Copy images
for filename in seedling["filename"]:
    source_path = os.path.join(source_folder, filename)
    dest_path = os.path.join(seedling_folder, filename)
    
    if os.path.exists(source_path):
        img = Image.open(source_path)
        img.save(dest_path)
    else:
        print(f"Not found: {filename}")

print("Seedling stage classification done!")
```

## Augmentation

```python
import os
import random
from PIL import Image, ImageEnhance, ImageOps

input_folder = "maize_growth_stage/flowering"

target_count = 1000

image_files = [f for f in os.listdir(input_folder) 
               if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

current_count = len(image_files)
print("Original images:", current_count)

count = 0

while current_count + count < target_count:
    
    img_name = random.choice(image_files)
    img_path = os.path.join(input_folder, img_name)
    
    img = Image.open(img_path)

    aug_type = random.choice([
        "flip", "rotate", "brightness", "contrast", "zoom"
    ])

    if aug_type == "flip":
        img = ImageOps.mirror(img)

    elif aug_type == "rotate":
        angle = random.randint(-30, 30)
        img = img.rotate(angle)

    elif aug_type == "brightness":
        enhancer = ImageEnhance.Brightness(img)
        img = enhancer.enhance(random.uniform(0.6, 1.4))

    elif aug_type == "contrast":
        enhancer = ImageEnhance.Contrast(img)
        img = enhancer.enhance(random.uniform(0.6, 1.4))

    elif aug_type == "zoom":
        width, height = img.size
        crop = img.crop((
            width*0.1, height*0.1,
            width*0.9, height*0.9
        ))
        img = crop.resize((width, height))

    new_name = f"aug_{count}_{img_name}"
    save_path = os.path.join(input_folder, new_name)
    
    img.save(save_path)
    count += 1

print("Augmentation complete!")
print("Total images:", len(os.listdir(input_folder)))
```
