```python
import os

folder_path = "Healthy"

image_extensions = ('.jpg', '.jpeg', '.png')

image_files = []

for file in os.listdir(folder_path):
    if file.lower().endswith(image_extensions):
        file_path = os.path.join(folder_path, file)
        size = os.path.getsize(file_path)
        image_files.append((file_path, size))

image_files.sort(key=lambda x: x[1])

print(f"Total images before deletion: {len(image_files)}")

remove_count = len(image_files) - 1000

for i in range(remove_count):
    os.remove(image_files[i][0])

print(f"Deleted {remove_count} images")
print("Remaining images: 1000")
```
