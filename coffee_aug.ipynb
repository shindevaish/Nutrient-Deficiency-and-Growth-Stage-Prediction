### Reducing healthy images
#### 1010 -> 1000

```python
import os

folder_path = "Healthy"

image_extensions = ('.jpg', '.jpeg', '.png')

image_files = []

for file in os.listdir(folder_path):
    if file.lower().endswith(image_extensions):
        file_path = os.path.join(folder_path, file)
        size = os.path.getsize(file_path)
        image_files.append((file_path, size))

image_files.sort(key=lambda x: x[1])

print(f"Total images before deletion: {len(image_files)}")

remove_count = len(image_files) - 1000

for i in range(remove_count):
    os.remove(image_files[i][0])

print(f"Deleted {remove_count} images")
print("Remaining images: 1000")
```

### Augmenting nitrogen disease plant (Total 1000)

```python
import os
import random
from PIL import Image, ImageOps

input_folder = "nitrogen-N"
output_folder = "nitrogen"

os.makedirs(output_folder, exist_ok=True)

image_files = [f for f in os.listdir(input_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

count = 0
for file in image_files:
    img = Image.open(os.path.join(input_folder, file))
    
    if img.mode in ("RGBA", "P"):
        img = img.convert("RGB")
        
    img.save(os.path.join(output_folder, f"orig_{count}.jpg"))
    count += 1

target = 1000

while count < target:
    file = random.choice(image_files)
    img = Image.open(os.path.join(input_folder, file))

    if img.mode in ("RGBA", "P"):
        img = img.convert("RGB")

    choice = random.choice(["rotate_left", "rotate_right", "flip"])

    if choice == "rotate_left":
        augmented = img.rotate(20)
    elif choice == "rotate_right":
        augmented = img.rotate(-20)
    elif choice == "flip":
        augmented = ImageOps.mirror(img)

    if augmented.mode in ("RGBA", "P"):
        augmented = augmented.convert("RGB")

    augmented.save(os.path.join(output_folder, f"aug_{count}.jpg"))
    count += 1

print(f"Total images created: {count}")
```
